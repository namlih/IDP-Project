```python
import pandas as pd
import numpy as np
from pylab import mpl, plt
import warnings
warnings.filterwarnings('ignore')
```

```python
plt.style.use('seaborn')
mpl.rcParams['font.family'] = 'serif'
%matplotlib inline
```

```python
from os import listdir
from os.path import isfile, join

path = "../data generators/generated_price_data/"
files = [f for f in listdir(path) if isfile(join(path, f))]
```

```python
files
```

```python
files_imp = ['btc-usd_daily.csv','eth-usd_daily.csv']
```

## Momentum with SMA Strategy

```python
for file in files_imp:
    #read file
    data = pd.read_csv("../data generators/generated_price_data/"+file)
    data = data[['market','day','close']]
    data.index = data['day']

    #take proportion of the data
    test_data = data[(data['day'] >= '2019-01-01')]

    #find buy&hold strategy returns
    test_data['returns'] = (test_data['close'] / test_data['close'].shift(1))-1
    test_data.dropna(inplace=True)
    test_data['buy&hold'] = (1+test_data['returns'].values).cumprod()

    smas = [7,14,30,60,120]
    for sma in smas:
        #find sma values
        test_data['sma'+str(sma)] = test_data['close'].rolling(sma).mean()
        test_data['sma'+str(sma)+'_shifted'] = test_data['sma'+str(sma)].shift(1)
        #test_data.dropna(inplace=True)

        #determine position
        col_name = 'momentum_sma_'+str(sma)
        test_data[col_name+'_position'] = np.where(test_data['sma'+str(sma)] > test_data['sma'+str(sma)+'_shifted'], 1, -1)

        #find returns    
        test_data[col_name+'_returns'] = test_data[col_name+'_position'].shift(1) * test_data['returns']
        test_data.dropna(inplace=True)

        #find cumulative returns
        test_data[col_name] = (1+test_data[col_name+'_returns'].values).cumprod()

    plot_columns = ['buy&hold']+['momentum_sma_'+str(sma) for sma in smas]

    plot_title = "Momentum with SMA Strategy "+file[:7]
    ax = test_data[test_data['day'] >= '2020-01-01'][plot_columns].plot(title=plot_title,figsize=(10, 6))
    ax.get_legend()
```
