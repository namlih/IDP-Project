```python
plt.style.use('seaborn')
mpl.rcParams['font.family'] = 'serif'
%matplotlib inline
```

## Data Import

```python
data = pd.read_csv("../data generators/generated_price_data/btc-usd_daily.csv")
data = data[['market','day','close']]
data.index = data['day']

trend = pd.read_csv("../data generators/generated_trends_data/bitcoin_daily.csv")
trend = trend[['keyword','day','trend']]
trend.index = trend['day']

pageview = pd.read_csv("../data generators/generated_wiki_pageview_data/Bitcoin_daily.csv")
pageview = pageview[['article','day','pageview']]
pageview.index = pageview['day']
```

```python
def sma_momentum_strategy(test_data,sma,column):
    #find sma values
    test_data['sma'+str(sma)] = test_data[column].rolling(sma).mean()
    test_data['sma'+str(sma)+'_shifted'] = test_data['sma'+str(sma)].shift(1)

    #determine position
    col_name = 'momentum_sma_'+str(sma)
    test_data[col_name+'_position'] = np.where(test_data['sma'+str(sma)] > test_data['sma'+str(sma)+'_shifted'], 1, -1)

    #find returns
    test_data['returns'] = (test_data['close'] / test_data['close'].shift(1))-1
    test_data[col_name+'_returns'] = test_data[col_name+'_position'].shift(1) * test_data['returns']
    test_data.dropna(inplace=True)
    
    test_data = test_data[test_data['day'] >= '2020-01-01']
    return (1+test_data[col_name+'_returns'].values).cumprod()

def trend_estimation_strategy(test_data,trend_length,column):
    #find trend estimations
    test_data['trend_estimation'] = (test_data[column] / test_data[column].shift(trend_length))-1

    #determine position
    col_name = 'trend_estimation'+str(trend_length)
    test_data[col_name+'_position'] = np.where(test_data['trend_estimation'] > 0, 1, -1)

    #find returns
    test_data['returns'] = (test_data['close'] / test_data['close'].shift(1))-1
    test_data[col_name+'_returns'] = test_data[col_name+'_position'].shift(1) * test_data['returns']
    test_data.dropna(inplace=True)

    #find cumulative returns
    test_data = test_data[test_data['day'] >= '2020-01-01']
    return (1+test_data[col_name+'_returns'].values).cumprod()
```

```python
#take proportion of the data
test_data = data[(data['day'] >= '2019-12-31')]

#find buy&hold strategy returns
test_data['returns'] = (test_data['close'] / test_data['close'].shift(1))-1
test_data.dropna(inplace=True)
test_data['buy&hold'] = (1+test_data['returns'].values).cumprod()
returns_data = test_data[['returns','buy&hold']].copy()

test_data = data[(data['day'] >= '2018-01-01')]
last_day = data['day'].values[-1]
test_data['trend'] = trend[(trend['day'] >= '2018-01-01') & (trend['day'] <= last_day)]['trend'].values
test_data['pageview'] = pageview[(pageview['day'] >= '2018-01-01') & (pageview['day'] <= last_day)]['pageview'].values

returns_data['momentum_sma14'] = sma_momentum_strategy(test_data.copy(),14,'close')
returns_data['momentum_sma120_GoogleTrends'] = sma_momentum_strategy(test_data.copy(),120,'trend')
returns_data['momentum_sma60_WikiPageview'] = sma_momentum_strategy(test_data.copy(),60,'pageview')

returns_data['momentum_trendestimation240'] = trend_estimation_strategy(test_data.copy(),240,'close')
returns_data['momentum_trendestimation120_GoogleTrends'] = trend_estimation_strategy(test_data.copy(),120,'trend')
returns_data['momentum_trendestimation240_WikiPageview'] = trend_estimation_strategy(test_data.copy(),240,'pageview')
```

```python
plot_columns = ['buy&hold',
                'momentum_sma14',
                'momentum_sma120_GoogleTrends',
                'momentum_sma60_WikiPageview',
                'momentum_trendestimation240',
                'momentum_trendestimation120_GoogleTrends']
                #'momentum_trendestimation240_WikiPageview']

ax = returns_data[plot_columns].plot(title="Bitcoin Baseline Strategies",figsize=(10, 6))
ax.get_legend()
```
